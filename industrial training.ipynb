```python
# Function for plotting bar graph for eachnumeric features, Univariate Analysis
def univariate_bar(col,ngo,color,place):
    
    if col == 'child_mort':
        word = "Child Mortality Rate"
    if col == 'exports':
        word = "Export"
    if col == 'health':
        word = "Health"
    if col == 'imports':
        word = "Imports"
    if col == 'income':
        word = "Income"   
    if col == 'inflation':
        word = "Inflation"
    if col == 'life_expec':
        word = "Life Expectation"
    if col == 'total_fer':
        word = "Fertility Rate"
    if col == 'gdpp':
        word = "GDPP"        
    
    fig = go.Figure()
    # Add Traces

    fig.add_trace(
        go.Bar(x=ngo['country'][:5],
               y=ngo[col][:5],
               name= place + " 5",
              marker={'color': ngo[col][:5],'colorscale': color}))
    fig.add_trace(
        go.Bar(x=ngo['country'][:10],
               y=ngo[col][:10],
               name=place + " 10",
              marker={'color': ngo[col][:10],'colorscale': color}))
    fig.add_trace(
        go.Bar(x=ngo['country'][:50],
               y=ngo[col][:50],
               name= place + " 50",
              marker={'color': ngo[col][:50],'colorscale': color}))
    fig.add_trace(
        go.Bar(x=ngo['country'][:100],
               y=ngo[col][:100],
               name= place + " 100",
              marker={'color': ngo[col][:100],'colorscale': color}))
    fig.add_trace(
        go.Bar(x=ngo['country'],
               y=ngo[col],
               name="All",
              marker={'color': ngo[col],'colorscale': color}))
    
    # Update the layout
    fig.update_layout(
        updatemenus=[
            dict(
                type="buttons",
                direction="left",
                active=0,
                x=0.8,
                y=1.6,
                buttons=list([
                    dict(label= place + " 5",
                         method="restyle",
                         args=[{"visible": [True, False,False, False, False]},
                               {"title": "Top 5 Countries with " + word }]),
                    dict(label= place + " 10",
                         method="restyle",
                         args=[{"visible": [False,True, False, False, False]},
                               {"title": "Top 10 Countries with " + word}]),
                    dict(label=place + " 50",
                         method="restyle",
                         args=[{"visible": [False,False, True, False, False]},
                               {"title": "Top 50 Countries with " + word}]),
                    dict(label= place + " 100",
                         method="restyle",
                         args=[{"visible": [False,False, False, True, False]},
                               {"title": "Top 100 Countries with " + word}]),
                    dict(label="All",
                         method="restyle",
                         args=[{"visible": [False,False, False, False, True]},
                               {"title": "All Countries"}]),
                ]),
            )
        ])
    fig.update_layout(
    title_text="Country Wise " + word,
    xaxis_domain=[0.05, 1.0]
    )
    fig.update_traces(showlegend=False)

    fig.show()
```

```python
#function for plotting the map for individual features vs country spread

def map_plot(col,colorscale):
    
    # Standardzing the title and button names
    if col == 'child_mort':
        word = "Child Mortality Rate"
    elif col == 'exports':
        word = "Export"
    elif col == 'health':
        word = "Health"
    elif col == 'imports':
        word = "Imports"
    elif col == 'income':
        word = "Income"   
    elif col == 'inflation':
        word = "Inflation"
    elif col == 'life_expec':
        word = "Life Expectation"
    elif col == 'total_fer':
        word = "Fertility Rate"
    elif col == 'gdpp':
        word = "GDPP"
    else:
        word = "Cluster"
        
    # Plotting the Map    
    data = [dict(
            type = 'choropleth',
            locations = ngo['country'],
            locationmode = 'country names',
            z = ngo[col],
            text = ngo['country'],
            colorscale = colorscale,
            autocolorscale = False,
            reversescale = False,
            marker_line_color='darkgray',
            marker_line_width=0.5,
            marker = dict(
                line = dict (
                    color = 'rgb(180,180,180)',
                    width = 0.5
                ) 
            ),
            colorbar = dict(
                autotick = False,
                title = word)
    )
           ]

    layout = dict(
        title = word + ' per Country',
        geo = dict(
            showframe = False,
            showcoastlines = True,
            projection = dict(
                type = 'Mercator')
        )
    )

    fig = dict(data=data, layout=layout)
    iplot(fig, validate=False)
```

Thus we will cap the <b>gdpp</b> , <b>health</b> and <b>income </b> features for the upper range only.</li>
            <li>For import & export, we will cap the upper limits as there are no lower side ourliers </li>
        </ul>
    </span>

```python
#finding the 99th percentile value to cap
upper_gdpp = ngo['gdpp'].quantile(0.99)
upper_income = ngo['income'].quantile(0.99)
upper_health = ngo['health'].quantile(0.99)
upper_import = ngo['imports'].quantile(0.99)
upper_export = ngo['exports'].quantile(0.99)
upper_inflation = ngo['inflation'].quantile(0.99)

print('Total number of rows getting capped for dgpp : ', len(ngo[ngo['gdpp']>upper_gdpp]))
print('Total number of rows getting capped for income : ', len(ngo[ngo['income']>upper_income]))
print('Total number of rows getting capped for health : ', len(ngo[ngo['health']>upper_health]))
print('Total number of rows getting capped for imports : ', len(ngo[ngo['imports']>upper_import]))
print('Total number of rows getting capped for exports : ', len(ngo[ngo['exports']>upper_export]))
print('Total number of rows getting capped for inflation : ', len(ngo[ngo['inflation']>upper_inflation]))


# capping the gdpp and income values
ngo['gdpp'][ngo['gdpp']>upper_gdpp] = upper_gdpp
ngo['income'][ngo['income']>upper_income] = upper_income
ngo['health'][ngo['health']>upper_health] = upper_health
ngo['imports'][ngo['imports']>upper_import] = upper_import
ngo['exports'][ngo['exports']>upper_export] = upper_export
ngo['inflation'][ngo['inflation']>upper_inflation]= upper_inflation
```

<h2>   
      <font color = purple >
            <span style='font-family:Georgia'>
            6.1 Hopkins Statistics
            </span>   
        </font>    
</h2>

<div class="alert alert-block alert-info">
    <span style='font-family:Georgia'>
        <b>Insight: </b><br>
        Before we apply any clustering algorithm to the given data, it's essential to check whether the given data has meaningful clusters. Which, in general means, the given data is not random. The process to evaluate the data to check if the data is feasible for clustering or not is know as the clustering tendency. To check cluster tendency, we use the Hopkins test. Hopkins test examines whether data points differ significantly from uniformly distributed data in the multidimensional space.
    </span>    
</div>

```python
def hopkins(X):
    d = X.shape[1]
    #d = len(vars) # columns
    n = len(X) # rows
    m = int(0.1 * n) 
    nbrs = NearestNeighbors(n_neighbors=1).fit(X.values)
 
    rand_X = sample(range(0, n, 1), m)
 
    ujd = []
    wjd = []
    for j in range(0, m):
        u_dist, _ = nbrs.kneighbors(uniform(np.amin(X,axis=0),np.amax(X,axis=0),d).reshape(1, -1), 2, return_distance=True)
        ujd.append(u_dist[0][1])
        w_dist, _ = nbrs.kneighbors(X.iloc[rand_X[j]].values.reshape(1, -1), 2, return_distance=True)
        wjd.append(w_dist[0][1])
 
    H = sum(ujd) / (sum(ujd) + sum(wjd))
    if isnan(H):
        print(ujd, wjd)
        H = 0
 
    return H
```

```python
#Use the Hopkins Statistic function by passing the above dataframe as a paramter
hopkins(ngo_scaled)
```

<div class="alert alert-block alert-info">
    <span style='font-family:Georgia'>
        <b>Insight: </b><br>
        A <b>Hopkins Statistic</b> value close to 1 tends to indicate the data is highly clustered, random data will tend to result in values around 0.5, and uniformly distributed data will tend to result in values close to 0.The Hopkins Statsitics for the current data is around 0.9 which is a good indicator that data seems to form good clusters.
    </span>    
</div>

```python
#Checking the scaled parameter dataframe
ngo_scaled.head()
```

```python
# elbow-curve/SSD
ssd = []
range_n_clusters = [2, 3, 4, 5, 6, 7, 8]
for num_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=num_clusters, max_iter=50)
    kmeans.fit(ngo_scaled)
    
    ssd.append(kmeans.inertia_)
    
# plot the SSDs for each n_clusters
ssddf = pd.DataFrame(ssd)
ssddf.columns = ['SSD']
fig = go.Figure(data=go.Scatter(x=ssddf.index, y=ssddf['SSD']))
fig.update_layout(height=500, width=800, title_text="SSD/Elbow Curve")
fig.show()
```

<h3>   
      <font color = darkgreen >
            <span style='font-family:Georgia'>
            6.3.2 Silhouette Analysis
            </span>   
        </font>    
</h3>

<span style="font-size:6mm" >
    <span style ='font-family:Georgia'>
        <font color = blue > Silhouette score =
            <span style="display: inline-block;vertical-align: middle;">
                <div style="text-align: center;border-bottom: 1px solid black;">p-q</div>
                <div style="text-align: center;">max(p,q)</div> 
            </span>
        </font>
    </span>
</span>

<p>
    <span style='font-family:Georgia'>
        <b>p</b>  is the mean distance to the points in the nearest cluster that the data point is not a part of <br>
        <b>q</b>  is the mean intra-cluster distance to all the points in its own cluster.
    </span>
</p>

<span style='font-family:Georgia'>
    <ul>
        <li>The value of the silhouette score range lies between -1 to 1.</li>
        <li>A score closer to 1 indicates that the data point is very similar to other data points in the cluster</li>
        <li>A score closer to -1 indicates that the data point is not similar to the data points in its cluster</li>
    </ul>
</span>
    
    

```python
# silhouette analysis
range_n_clusters = [2, 3, 4, 5, 6, 7, 8]

for num_clusters in range_n_clusters:
    
    # intialise kmeans
    kmeans = KMeans(n_clusters=num_clusters, max_iter=50)
    kmeans.fit(ngo_scaled)
    
    cluster_labels = kmeans.labels_
    
    # silhouette score
    silhouette_avg = silhouette_score(ngo_scaled, cluster_labels)
    print("For n_clusters={0}, the silhouette score is {1}".format(num_clusters, silhouette_avg))
```

```python
sse_ = []
for k in range(2, 15):
    kmeans = KMeans(n_clusters=k).fit(ngo_scaled)
    sse_.append([k, silhouette_score(ngo_scaled, kmeans.labels_)])


plt.plot(pd.DataFrame(sse_)[0], pd.DataFrame(sse_)[1])
plt.show()
```

    <span style='font-family:Georgia'>
